```python
import fnmatch

xlsx_names = []
for i in os.listdir("./reviews"):
    if fnmatch.fnmatch(i,"*.xlsx"):
        xlsx_names.append("./reviews/" + i) 
```

```python
def get_reviews(file_name,sheet_index = 0,col_num = 2):
    data = xlrd.open_workbook(file_name)
    review = data.sheet_by_index(sheet_index).col_values(col_num)
    return review
```

```python
# 每篇文章的评论
reviews = []
for i in xlsx_names:
    reviews.append(get_reviews(i))
```

```python
# 所有文章的评论
reviews_all = []
for i in xlsx_names:
    reviews_all.extend(get_reviews(i))
```

```python
from collections import Counter
```

```python
import jieba.posseg as pseg
```

```python
word_freq_all = Counter()
allowPOS = ('ns','n', 'a', 'ad')
for review in reviews_all:
    words = pseg.cut(review)
    for word,flag in words:
        if flag in allowPOS:
            word_freq_all[word]+=1
```

```python
for word, cnt in word_freq_all.most_common(320):
    print word,cnt
```
